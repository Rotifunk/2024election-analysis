```python
nltk.download('stopwords')
```

### stopwords 정리, 토크나이즈

```python
def text_prep(text_list):
    stop_words = set(stopwords.words('english'))
    stop_words.update(['said', 'could', 'would', 'also', 'even', 'mr', 'president'])
    word_token = []
    for contents in text_list:
        cleaned_text = re.sub('[^a-zA-Z ]', '', contents)
        words = word_tokenize(cleaned_text)
        for word in words:
            if word.lower() not in stop_words:
                word_token.append(word)
    return word_token
```

```python
# csv파일 일기
news_df = pd.read_csv('../collect_data/result_7200.csv').drop(columns=['Unnamed: 0']).dropna()
```

### wordcloud 생성

```python
data = text_prep(news_df[news_df['section']=='us']['text'])
wordCloud = WordCloud(
    width=400,
    height=400,
    max_font_size=100,
    background_color='white',
)
wordCloud.generate(str(data).lower())
wordCloud.to_file('test.jpg')
```
